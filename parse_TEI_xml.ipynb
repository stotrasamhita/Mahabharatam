```python
from bs4 import BeautifulSoup
import re
```

```python
mb_text = []
with open('SARIT-corpus/mahabharata-devanagari.xml') as mb_file:
    mb_text = mb_file.read()
```

```python
soup = BeautifulSoup(mb_text, 'lxml')
```

```python
div_list_all = soup.find_all('div')
```

```python
parva_num = 0
for i, div in enumerate(div_list_all):
    if div.find_all('div') != []:
        parva_num += 1
        chap_num = 0
    else:
        chap_num += 1

    if chap_num == 0:
        continue
        
    fname = 'parvas/%02d-%03d.md' % (parva_num, chap_num)
    
    with open(fname, 'w') as out_file:
#         for p in div.find_all('p'):
#             if 'topic' in p.get_attribute_list('rend'):
#                 out_text = re.sub(r'\s+',' ', p.text)
#                 print('### Topics\n', out_text, file=out_file)
#             elif 'footnotes' in p.get_attribute_list('rend'):
#                 out_text = re.sub(r'\s+',' ', p.text)
#                 print ('### Footnotes\n', out_text, file=out_file)
#             elif p.text.split()[0] == '':
#                 out_text = re.sub(r'\s+',' ', p.text)
#                 print(out_text, file=out_file)
#                 print('%%%% Here ends the Adhyaya!', file=out_file)
#             else:
#                 print(chap_num, file=out_file)
#                 print('%%%% Here begins the Adhyaya %03d!' % chap_num, file=out_file)
#                 print('## ', p.text, file=out_file)
                

        snum = 0
        print ('## Div %03d' % i, file=out_file)
        for lg in div.find_all('lg'):
            for l in lg.find_all('l'):
                for seg in l.find_all('seg'):
                    if 'a' in seg.get_attribute_list('n'):
                        snum += 1
                    out_text = re.sub('\s+', ' ', seg.text)
                    print('%03d%s\t%s' % (snum, seg.get_attribute_list('n')[0], out_text), file=out_file)
```
