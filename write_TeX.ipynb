```python
for json_file in os.listdir("extracted_data"):
  with open("extracted_data/%s" % json_file, encoding='utf-8') as json_data:
    data = json.load(json_data, strict=False)
    parva_name = json_file[3:-5]
    print(parva_name)
    data = data[parva_name]
    with open('parvas/'+json_file.replace(".json",".tex"), 'w') as parva_file:
        print ('\part{%s}' % sanscript.transliterate(parva_name, _from="iso", _to="devanagari"), file=parva_file)
        for i, adhyaya in enumerate(data):
          print ('\chapter{अध्यायः %s}' % sanscript.transliterate(str(i+1), _to="devanagari"), file=parva_file)
          for verse in data[adhyaya]['verses']:
            shloka = verse['verse_text']
            shloka = [s.strip("\xa0॥ ") for s in shloka]
            if len(shloka) == 2:
              print('\\twolineshloka\n{%s}\n{%s}\n\n' % tuple(shloka), file=parva_file)
            elif len(shloka) == 4:
              print('\\fourlineindentedshloka\n{%s}\n{%s}\n{%s}\n{%s}\n\n' % tuple(shloka), file=parva_file)
            elif len(shloka) == 3:
              print('\\threelineshloka\n{%s}\n{%s}\n{%s}\n\n' % tuple(shloka), file=parva_file)
            else:
              print('%% Check verse!\n%s' % '\n\n'.join(shloka), file=parva_file)
```
